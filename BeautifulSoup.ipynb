```python
import requests
import fake_useragent
from bs4 import BeautifulSoup
```

```python
# Get User-Agent
ua = fake_useragent.UserAgent()
header = {"User-Agent": ua.random}

# Scrape web source code
url = "https://www.gaoqingw.com/24635.html"
response = requests.get(url, headers = header)
html = response.text
print(html)

re_obj = BeautifulSoup(html, 'html.parser')
```

```python
# Finds the element within the specified tag name
attr = {"class": "ico_2"}

urls = re_obj.find_all('a', attrs = attr)  # ('Tag name', filter condition)

for i in urls:
    print(i.get("title"))  # episode
    print(i.get('href'))  # link
```
