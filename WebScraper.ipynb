```python
import re
import requests
import fake_useragent
```

```python
# Get User-Agent
ua = fake_useragent.UserAgent()
header = {"User-Agent": ua.random}

# re.compile(pattern, flags = 0)
re_obj = re.compile('<em class="">(.*?)</em>.*?<span class="title">(.*?)</span>.*?<span class="title">&nbsp;/&nbsp;.*?(.*?)</span>.*?<span class="inq">(.*?)</span>', re.S)

for i in range(10):
    # define URL
    url = "https://movie.douban.com/top250?start={0}&filter=".format(i*25)
    # send request
    response = requests.get(url, headers = header)
    # Get response
    html = response.text
    film_list = re_obj.findall(html)
    for film in film_list:
        print(film)
        
response.close()
```

```python
import re
import jieba
```

```python
stopwords = [i.strip() for i in open('hit_stopwords.txt',encoding = 'UTF-8').readlines()]
text = ''
for str in data:
    # spilit words
    cut_list = jieba.lcut(''.join(re.findall('[\u4e00-\u9fa5]',str)))  #Keep only Chinese words
    for word in cut_list:
        if word not in stopwords:
            text += ' ' + word
text
```
